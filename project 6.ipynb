```python
import pandas as pd

# Sample Dataset
data = {
    "CustomerID": [101,102,103,104,105,106,107,108,109,110],
    "Age": [22,25,35,28,40,31,27,45,38,29],
    "MonthlyCharges": [500,700,600,800,900,750,650,1000,850,720],
    "Tenure": [12,24,18,36,48,30,20,60,40,25],
    "Churn": ["Yes","No","Yes","No","No","Yes","Yes","No","No","Yes"]
}

df = pd.DataFrame(data)

print("===== DATASET PREVIEW =====")
print(df)

print("\nDataset Shape:", df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nTarget Variable (Churn):")
print(df["Churn"].value_counts())

print("\n===== ML FEASIBILITY ASSESSMENT =====")

rows, cols = df.shape

if rows >= 10 and "Churn" in df.columns:
    print("Dataset Size: Suitable for learning purpose.")
    print("Target Label Found: Churn")
    print("Recommended ML Approach: Classification")
    print("Conclusion: Machine Learning is feasible for this dataset.")
else:
    print("Dataset is not suitable for Machine Learning.")

print("\nBusiness Justification:")
print("- Customer churn can be predicted using historical customer data.")
print("- A classification model can identify customers who may leave.")
print("- This helps the business improve customer retention.")
```
